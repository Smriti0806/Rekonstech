```python
import pandas as pd

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_data = pd.read_csv(url)

# Display basic information about the dataset
print("Basic Information:")
print(titanic_data.info())

# Display summary statistics of numerical columns
print("\nSummary Statistics:")
print(titanic_data.describe())

# Display the first few rows of the dataset
print("\nFirst few rows:")
print(titanic_data.head())

# Perform specific analysis tasks
# Example: Calculate the survival rate
survival_rate = titanic_data['Survived'].mean() * 100
print("\nSurvival Rate:", "{:.2f}%".format(survival_rate))

# Example: Calculate the average age of passengers
average_age = titanic_data['Age'].mean()
print("Average Age:", "{:.2f}".format(average_age))

# Example: Count the number of male and female passengers
gender_counts = titanic_data['Sex'].value_counts()
print("\nNumber of Passengers by Gender:")
print(gender_counts)

# Example: Calculate the average fare paid by passengers
average_fare = titanic_data['Fare'].mean()
print("\nAverage Fare:", "{:.2f}".format(average_fare))
```
