```python
#Task 4: String Manipulation
# Concatenation
str1 = "Hello"
str2 = "World"
concatenated_str = str1 + " " + str2
print("Concatenation result:", concatenated_str)

# Slicing
original_str = "Python is awesome"
sliced_str = original_str[7:10]  # Extracting "is "
print("Slicing result:", sliced_str)

# Formatting
name = "John"
age = 30
formatted_str = "My name is {} and I am {} years old.".format(name, age)
print("Formatting result:", formatted_str)

# String Interpolation
name = "Alice"
age = 25
interpolated_str = f"My name is {name} and I am {age} years old."
print("String Interpolation result:", interpolated_str)
```
